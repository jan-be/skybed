```python
import re
import pandas as pd
from matplotlib import pyplot as plt
from scipy.stats import alpha

plt.rcParams['figure.dpi'] = 600
```

```python
# Define the regex pattern to capture integer or decimal time values
pattern = re.compile(r'time=(\d+(?:\.\d+)?)\s+ms')

# Function to extract ping times from a file and label with the file source
def extract_ping_times(filename):
    time_values = []
    
    # Open the file and extract time values
    with open(filename, 'r') as file:
        for line in file:
            match = pattern.search(line)
            if match:
                time_values.append(float(match.group(1)))

    return time_values


# Create a DataFrame with columns for time and source
df = pd.DataFrame({'Simulated': extract_ping_times("simulated.txt"), 'Precalculated': extract_ping_times("precalculated.txt")})

df
```

```python
df.plot(y=["Simulated", "Precalculated"], xlabel="Time (s)", ylabel="Ping Time (ms)", alpha=0.75)
```
